#### Binary classification of two Iris classes with Logistic Regression.

Let's load the Iris dataset using the sklearn module.

```python
data = load_iris()
```

```python
data.feature_names
```

```python
data.target_names
```

```python
data.target
```

```python
full_data = np.concatenate((data.data,data.target.reshape(-1,1)),axis=1)
df = pd.DataFrame(full_data, columns=data.feature_names+["Species"])
df.drop(df[df['Species'] == 2].index,inplace=True)
df.drop(['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)'], axis=1, inplace=True)
df
```

Logistic Regression using sklearn module

```python
X = np.array(df['petal width (cm)']).reshape(-1,1)
y = np.array(df['Species'], dtype=int)
```

```python
# Split the data into training (80%) and testing (20%) datasets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)

print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)

print(f'Number of {data.target_names[1]} samples in the data: {sum(y)} out of {len(y)}')
print(f'Number of {data.target_names[1]} samples in the train dataset: {sum(y_train)} out of {len(y_train)}')
print(f'Number of {data.target_names[1]} samples in the test dataset: {sum(y_test)} out of {len(y_test)}')
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)

y_pre_train_sk = model.predict(X_train)
y_pre_test_sk = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Confusion matrix Score of our model on the test dataset
confusion_matrix(y_test, y_pre_test_sk)
```

```python
# Accuracy Score of Sklearn model: fraction of correctly classified samples
accuracy_score(y_test,y_pre_test_sk)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pre_test_sk))
```

```python
X_new = np.linspace(0,2,2000).reshape(-1,1)
y_pre_new_sk = model.predict_proba(X_new)

sb.scatterplot(x="petal width (cm)", y="Species", data=df, hue='Species', palette=['red', 'blue'])
plt.plot(X_new,y_pre_new_sk[:,1],"g")
plt.show()
```

---
#### ROC curve

```python
from sklearn import metrics

y_test_pred_prob = model.predict_proba(X_test)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_test_pred_prob)
auc = metrics.roc_auc_score(y_test, y_test_pred_prob)
```

```python
# Plot ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend()
plt.show()
```

What does metrics.roc_curve do?

```python
thresholds = np.linspace(0,1,100)
fpr = []
tpr = []
for thr in thresholds:
    y_pre_test = 1 * (y_test_pred_prob > thr)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pre_test).ravel()
    fpr.append(fp/(fp+tn))
    tpr.append(tp/(fn+tp))
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.plot(thresholds,fpr)
plt.xlabel("Threshold")
plt.ylabel('False Positive Rate')

plt.subplot(1, 2, 2)
plt.plot(thresholds,tpr)
plt.xlabel("Threshold")
plt.ylabel('True Positive Rate')

plt.show()
```

#### "Manual" fitting of the logistic regression model: gradient descent in detail

We want to maximise $l(\beta)$, thus minimise $-l(\beta)$ (see slides). We calculate the gradient of $-l(\beta)$, which is given by the following equations:

$$
\frac{\partial (-l(\beta))}{\partial \beta_0} = \sum_{i=1}^{N}(p(x_i;\beta) - y_i)
$$

and for $k=1,2,\ldots,k$:

$$
\frac{\partial (-l(\beta))}{\partial \beta_0} = \sum_{i=1}^{N}x_{ij}(p(x_i;\beta) - y_i).
$$

```python
def gradient_descent(x, y, epochs, learning_rate):
    
    num_samples, num_features = x.shape
    # Random value to weights
    beta = np.zeros(num_features)
    beta0 = 0
    for i in range(epochs):
        # Sigmoid Function
        p_predict = 1/(1 + np.exp( -(x.dot(beta)+beta0) ))
        # Gradient Calculation
        diff = p_predict - y.T
        diff = np.reshape(diff, num_samples)
        
        dbeta = np.dot(x.T, diff)
        dbeta0 = np.sum(diff)
        
        #dW = np.dot(x.T, diff)/num_samples
        #db = np.sum(diff)/num_samples
        
        # Updating Weights
        beta = beta - learning_rate * dbeta
        beta0 = beta0 - learning_rate * dbeta0
    
    return beta, beta0

def predict(x, beta, beta0, thr=0.5):
    p = 1/(1+ np.exp( -(x.dot(beta)+beta0) )) # gives probability
    y = np.where(p > thr, 1, 0)               # threshold value is thr, by default set to 0.5 
    return y, p

# Model training: fitting beta and beta0 with gradient descent
beta, beta0 = gradient_descent(X_train, y_train, 1000, 0.01)

y_pre_scratch, p_pre_scratch = predict(X_test, beta, beta0)
```
